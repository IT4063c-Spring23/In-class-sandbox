# Model & Notations

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 11)

def plot_line(y, color):
  y_ = eval(y)
  plt.plot(x, y_, label=f'y={y}', marker='o', markersize=5, color=color)
  plt.legend(loc='best')
  plt.xlabel('x')
  plt.ylabel('y')
```

```python
plot_line(y='x', color='red')
plot_line(y='3*x+5', color='blue')
plot_line(y='-2*x+4', color='green')
plot_line(y='2*x', color='purple')

plt.grid()
plt.show()
```

```python
fig = plt.figure()

ax = fig.add_subplot(111,projection='3d')

x1, x2 = np.meshgrid(range(10), range(10))

y_hat = 9 - x1 - x2

ax.plot_surface(x1, x2, y_hat, alpha=0.5)

plt.show()
```

# Multiple Linear Regression Using Scikit-Learn

```python
import pandas as pd

df2 = pd.read_csv('./data/test.csv')
df2
```

```python
# Visualize the data
import plotly.express as px

fig = px.scatter_3d(df2, x='x', y='y', z='z')
```

```python
from sklearn.linear_model import LinearRegression

# Train the model
model4 = LinearRegression()
model4.fit(df2[['x', 'y']], df2['z'])

# Make Predictions
```

```python
import numpy as np
import plotly.graph_objects as go

# Visualize the model
## Meshgrid
x_surf, y_surf = np.meshgrid(
  np.linspace(df2['x'].min(), df2['x'].max(), 100),
  np.linspace(df2['y'].min(), df2['y'].max(), 100)
)
## What does the meshgrid look like?
# x_surf = np.arange(df2['x'].min(), df2['x'].max(), 3)
# y_surf = np.arange(df2['y'].min(), df2['y'].max(), 3)
print(x_surf)
print(x_surf.shape)

# Make a dataframe from the meshgrid
X_surface = pd.DataFrame({'x': x_surf.ravel(), 'y': y_surf.ravel()})
print(X_surface)
# print(X_surface.shape)

# Make predictions z^
z_hat = model4.predict(X_surface)

fig.add_traces(
  go.Surface(
    x=x_surf,
    y=y_surf,
    z=z_hat.reshape(x_surf.shape),
    opacity=0.5,
    showscale=False
  )
)
```

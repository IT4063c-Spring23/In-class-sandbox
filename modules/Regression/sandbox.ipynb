# Model & Notations

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 11)

def plot_line(y, color):
  y_ = eval(y)
  plt.plot(x, y_, label=f'y={y}', marker='o', markersize=5, color=color)
  plt.legend(loc='best')
  plt.xlabel('x')
  plt.ylabel('y')
```

```python
plot_line(y='x', color='red')
plot_line(y='3*x+5', color='blue')
plot_line(y='-2*x+4', color='green')
plot_line(y='2*x', color='purple')

plt.grid()
plt.show()
```

```python
fig = plt.figure()

ax = fig.add_subplot(111,projection='3d')

x1, x2 = np.meshgrid(range(10), range(10))

y_hat = 9 - x1 - x2

ax.plot_surface(x1, x2, y_hat, alpha=0.5)

plt.show()
```

# Simple Linear Regression Using Scikit-Learn

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.DataFrame([
    [1  , 1], 
    [1.5, 2.3], 
    [2.5, 2.1], 
    [3  , 3], 
    [4  , 2.9]
  ], columns=['x', 'y'])

df
```

## Training the Model

```python
# Create a model
model = LinearRegression()



# model.fit(..., ...)
# what type of data does the model expect?
```

```python
# df['x'] is a pandas Series
# df[['x']] is a pandas DataFrame
print(df['x'])
print(df['x'].shape)
print(type(df['x']))

print(df[['x']])
print(df[['x']].shape)
print(type(df[['x']]))
```

```python
# Train the model
model.fit(df[['x']], df['y'])
```

```python
# Get the coefficients
print(model.coef_)
print(model.intercept_)
```

## Using the Model: Making Predictions
[`.predict()`](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html#sklearn.linear_model.LinearRegression.predict)

```python
# Make predictions for the x values of 0, and 1
data_to_predict = pd.DataFrame({
    'x': [20, 36]
})
# print(model.predict([[0]]))
# print(model.predict([[1]]))
print(model.predict([[0],[1]]))
print(model.predict(data_to_predict))


# explain the warning message
```

### Score
[`.score()`](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html#sklearn.linear_model.LinearRegression.score)

```python
model.score(df[['x']], df['y'])
```

## Visualize the model

```python
x1 = np.linspace(
    df['x'].min(),
    df['x'].max(),
    4
)
x1_reshaped = x1.reshape(-1, 1)
print(x1_reshaped)
y1 = model.predict(x1_reshaped)
y1
```

```python
x1
# [[1], [2.5], [4]]
```

```python
plt.scatter(df['x'], df['y'])
plt.plot(x1, y1)
plt.show()
```

# Simple Polynomial Regression Using Scikit-Learn
single feature - higher order/degree

## Manual Polynomial Features

```python
# Prepare the data
df['x_squared'] = df['x'] ** 2
df

# Train the model
model2 = LinearRegression()
model2.fit(df[['x', 'x_squared']], df['y'])

# Make Predictions
model2.predict([[4, 16]])
```

```python
# Visualize the model
x2 = np.linspace(
    df['x'].min(),
    df['x'].max(),
    100
)
x2_reshaped = x2.reshape(-1, 1)
x2_squared = x2_reshaped ** 2

new_x2 = np.hstack([x2_reshaped, x2_squared])

y2 = model2.predict(new_x2)
y2
```

```python
plt.scatter(df['x'], df['y'])
plt.plot(x2, y2)
plt.show()
```

## Automatic Polynomial Features

```python
# Prepare the data
poly = PolynomialFeatures(degree=5)
poly_X = poly.fit_transform(df[['x']])
poly_X

# Train the model
model3 = LinearRegression()
model3.fit(poly_X, df['y'])

# Make Predictions
```

```python
# Visualize the model
x3 = np.linspace(
    df['x'].min(),
    df['x'].max(),
    100
)
y3 = model3.predict(
    poly.fit_transform(x3.reshape(-1,1))
)
plt.scatter(df['x'], df['y'])
plt.plot(x3, y3)
plt.show()
```
